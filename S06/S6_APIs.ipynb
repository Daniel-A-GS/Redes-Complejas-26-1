````python
# Biblioteca para realizar solicitudes HTTP
import requests
# Manipulacin de datos
import pandas as pd
# Imprime estructuras de datos complejas de manera legible
import pprint
# Crea mapas interactivos
import folium
# Decodifica polilneas en formato FlexPolyline
import flexpolyline as fp
# Visualizacin de datos estadsticos
import seaborn as sns
import matplotlib.pyplot as plt
# Tiempo
import pytz
from datetime import datetime
````

````python
def utctodate(utc):
    # Obtener la zona horaria local
    local_tz = pytz.timezone('America/Mexico_City')

    # Convertir el timestamp a datetime
    utc_time = datetime.fromtimestamp(utc, pytz.utc)

    # Convertir de UTC a la zona horaria local
    local_time = utc_time.astimezone(local_tz)

    # Formatear la fecha y hora
    formatted_time = local_time.strftime('%Y-%m-%d %H:%M:%S')
    return formatted_time

def draw_map(df):
  m = folium.Map(location=(19.416795,	-99.192508), zoom_start=15)
  for i, item in df.iterrows():
      folium.Marker(
          location=[item['lat'], item['lon']],
          popup=f"{item['name']}, {item['lat']}, {item['lon']}"
      ).add_to(m)
  return m
````

````python
# Define la URL donde se encuentra el archivo JSON con la informacin
# de las estaciones de bicicletas
URL = 'https://gbfs.mex.lyftbikes.com/gbfs/en/station_information.json'

#Realiza una solicitud GET a la URL en formato JSON
r = requests.get(URL)

#Extrae la informacin de las estaciones
stations_data = r.json()['data']['stations']

df_stations = pd.DataFrame(stations_data)
df_stations
````

Mapa de las estaciones

````python
m = folium.Map(location=(19.416795,	-99.192508), zoom_start=15)
for i, item in df_stations.iterrows():
    folium.Marker(
        location=[item['lat'], item['lon']],
        popup=f"{item['name']}, {item['lat']}, {item['lon']}"
    ).add_to(m)
m
````

````python
#Estado de las estaciones de bicicletas
URL = 'https://gbfs.mex.lyftbikes.com/gbfs/en/station_status.json'

r = requests.get(URL)
df_status_stations = pd.DataFrame(r.json()['data']['stations'])

#Fecha ultima de actualizacin
print(utctodate(r.json()['last_updated']))

df_status_stations
````

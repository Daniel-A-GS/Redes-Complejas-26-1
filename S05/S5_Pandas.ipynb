# Conociendo el Dataset

Este conjunto de datos incluye información para la estimación de los niveles de obesidad en individuos de los países de:
- México
- Perú
- Colombia

basada en sus hábitos alimenticios y condición física. Los datos contienen 17 atributos y 2111 registros, los cuales están etiquetados con la variable de clase `NObesity` (Nivel de Obesidad), que permite la clasificación de los datos utilizando los valores: **Peso Insuficiente, Peso Normal, Sobrepeso Nivel I, Sobrepeso Nivel II, Obesidad Tipo I, Obesidad Tipo II y Obesidad Tipo III.** El **77%** de los **datos** fue generado de manera **sintética** utilizando la herramienta Weka y el filtro SMOTE, y el **23%** de los datos fue recolectado directamente de los **usuarios** a través de una plataforma web.

## Modificar un DataFrame

```python
#pop columna(s) del df y la devuelve
gender_series = df.copy().pop("Gender")
gender_series.head()
```

```python
#query (filtrado con condiciones)
print("Personas con malos hábitos")
x = df.query("BMI > 25 and family_history_with_overweight == 'yes' and SMOKE == 'yes'")
print(x.shape)
x
```

```python
#apply
df["Age"] = df["Age"].apply(round)
df
```

```python
#map (solo para series)
df['SMOKE_NUM'] = df['SMOKE'].map({'yes': 1, 'no': 0})
df.head(3)
```

```python
df["Age_squared"] = df["Age"].map(lambda x: x**2)
df.head(2)
```

```python
df.drop(columns="Age_squared", inplace=True)
```

```python
#pipe
def bmi_category(df):
    df["BMI_Category"] = pd.cut(df["BMI"], bins=[0,18.5,25,30,100],
                                labels=["Bajo peso","Normal","Sobrepeso","Obesidad"])
    return df

df = df.pipe(bmi_category)
df.head()
```

```python
df['BMI_Category'].unique()
```

```python
#no fumen
def drop_smokers(df): return df[df["SMOKE"] == "no"]
#hagan mucho ejercicio
def filter_FAF(df): return df[df["FAF"] >= 3]
#no tomen alcohol
def filter_CALC(df): return df[df["CALC"] == "no"]

df_healthy = (df
    .pipe(drop_smokers)
    .pipe(filter_FAF)
    .pipe(filter_CALC)
)
print(df_healthy.shape)
df_healthy
```

## Melt

Convertir columnas a filas

Pasar de formato **ancho** a **largo**

El formato largo es mejor para:
- visualizaciones (graficar muchas variables)
- procesamiento de datos para modelos de ML

`id_vars`: Columnas que no se "derretirán".

`var_name`: El nombre de la nueva columa que guarda los nombres originales de las columnas.

`value_name`: El nombre de la nueva columna que guarda los valores

`ignore_index`: Si False mantiene el índice original

¿Cuál es la distribución de los diferentes hábitos de consumo (frecuencia de consumo de vegetales, consumo de agua, actividad física, etc.) entre los sujetos?

```python
import matplotlib.pyplot as plt
import seaborn as sns

habitos_melted = df.melt(
    id_vars=['Gender', 'Age', 'MTRANS'],        #Columnas que se mantienen
    value_vars=['FCVC', 'CH2O', 'FAF', 'TUE'],  #Columnas de hábitos
    var_name='Tipo_Habito',           #Nombre para la columna de tipos
    value_name='Valor'                #Nombre para la columna de valores
)
print(habitos_melted.shape)
habitos_melted.head()
```

```python
habitos_pivoted = habitos_melted.pivot_table(
    index=['Gender', 'Age', 'MTRANS'],
    columns='Tipo_Habito',
    values='Valor',
    aggfunc='mean'
).reset_index()

print(habitos_pivoted.shape)
habitos_pivoted.head()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=habitos_melted, x='Tipo_Habito', y='Valor', hue='MTRANS', errorbar=None)
plt.xlabel('Hábito')
plt.ylabel('Valor del hábito')
plt.legend(title='Género')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=habitos_melted, x='Tipo_Habito', y='Valor', hue='Gender', errorbar=None)
plt.xlabel('Hábito')
plt.ylabel('Valor del hábito')
plt.legend(title='Género')
plt.show()
```

```python
#Obtener var dummies
df_encoded = df.copy()
#Saber quien es que valor
category_mappings = {}

#Columnas a convertir en dummies
for col in ["family_history_with_overweight", "CAEC", "SMOKE", "CALC", "MTRANS"]:
    #Convertir a categórica
    df_encoded[col] = df_encoded[col].astype("category")
    #Guardar en el diccionario el mapping del encoding
    category_mappings[col] = dict(enumerate(df_encoded[col].cat.categories))
    #Convertir a dummies los datos "categorizados"
    df_encoded[col] = df_encoded[col].cat.codes

print(category_mappings["MTRANS"])
```
